```python
import tensorflow as tf
import matplotlib.pyplot as plt

# Define a function to plot the feature maps from convolutional layers
def visualize_feature_maps(model, image, layer_name):
    # Define a model that outputs the feature maps from the selected layer
    layer_model = tf.keras.models.Model(inputs=model.input, outputs=model.get_layer(layer_name).output)

    # Get feature maps for the input image
    feature_maps = layer_model.predict(image)

    # Plot the feature maps
    num_feature_maps = feature_maps.shape[-1]
    plt.figure(figsize=(12, 12))

    for i in range(num_feature_maps):
        plt.subplot(8, 8, i+1)  # Adjust grid size for your feature map
        plt.imshow(feature_maps[0, :, :, i], cmap='viridis')
        plt.axis('off')

    plt.show()

# Example usage: Visualizing feature maps from a layer
image = np.expand_dims(image, axis=0)  # Add batch dimension
visualize_feature_maps(model, image, 'conv2d')  # 'conv2d' is the name of a convolutional layer
```
