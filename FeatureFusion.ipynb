```python
def fuse_features(image, nir_band, red_band):
    # Extract Pixel Intensity Histogram
    hist = extract_histogram(image)

    # Extract GLCM Texture Features
    contrast, correlation, energy, homogeneity = extract_glcm_features(image)

    # Calculate NDVI
    ndvi = extract_ndvi(image, nir_band, red_band)

    # Combine all features into one vector
    combined_features = np.concatenate([hist,
                                       [contrast, correlation, energy, homogeneity],
                                       [ndvi]], axis=0)
    return combined_features

# Example usage
image = cv2.imread('landslide_image.jpg')
nir_band = cv2.imread('nir_band.jpg', 0)
red_band = cv2.imread('red_band.jpg', 0)
features = fuse_features(image, nir_band, red_band)
print(f"Combined Features: {features}")
```
