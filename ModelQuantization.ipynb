```python
import tensorflow as tf

# Load the trained model
model = load_model('landslide_detection_model.h5')

# Convert model to TensorFlow Lite format with quantization
converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations = [tf.lite.Optimize.DEFAULT]  # Apply default optimizations
tflite_model = converter.convert()

# Save the TFLite model
with open('landslide_detection_model_quantized.tflite', 'wb') as f:
    f.write(tflite_model)

# Run inference on a TFLite interpreter (example for Raspberry Pi)
interpreter = tf.lite.Interpreter(model_path="landslide_detection_model_quantized.tflite")
interpreter.allocate_tensors()

# Set input tensor
input_details = interpreter.get_input_details()
interpreter.set_tensor(input_details[0]['index'], X_test)

# Run inference
interpreter.invoke()

# Get output tensor
output_details = interpreter.get_output_details()
output_data = interpreter.get_tensor(output_details[0]['index'])
print(output_data)
```
